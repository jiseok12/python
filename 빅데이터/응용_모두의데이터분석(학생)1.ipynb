```python
# 동그란 원 그리기 : axis()함수 이용
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic') 
plt.rcParams['axes.unicode_minus']=False

size = [2441,2312,1032,1233]
label = ['A형','B형','AB형','O형']
plt.axis('equal')
plt.pie(size, labels=label)
plt.show()
```

```python
# 동그란 원 그리기 : axis()함수 이용
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic') 
plt.rcParams['axes.unicode_minus']=False

size = [2441,2312,1032,1233]#데이터 값
label = ['A형','B형','AB형','O형'] #matplotlib 사용할수 있는 다양한 색의 이름
color = ['darkmagenta','deeppink','hotpink','pink']

plt.axis('equal')#동그란 원 옵션
plt.pie(size, labels=label,autopct='%.1f%%',colors=color, explode=(0,0,0.1,0)) #label 옵션을 통해 레이블 사용
plt.show()
```

```python
filename =filename='/content/drive/MyDrive/DATA/gender.csv'

import csv

f = open(filename, 'r', encoding='cp949')
data = csv.reader(f)
loc_name = '부산광역시 연제구 연산제1동'
m_list = 0
f_list = 0
for row in data:
  if city in row[0]:
    for m in row[3:104]:
      m_list=m_list+int(m.replace(',',''));
    for fe in row[106:]:
      f_list=f_list+int(fe.replace(',',''));
f.close()
value=[m_list,f_list]
print(f'남성의 총인구수:{m_list}')
print(f'여성의 총인구수:{f_list}')

label=['남성의 총인구수','여성의 총인구수']
color=['skyblue','orange']
plt.axis('equal')#동그란 원 옵션
plt.pie(value, labels=label,autopct='%.1f%%',colors=color,startangle=90) #label 옵션을 통해 레이블 사용
plt.show()
```
