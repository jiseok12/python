# 4)집합
* 특징
  * 중복없는 요소로만 구성
  * 중괄호로 작성되며 각 요소는 쉼표로 구분
  * 순서 없음
* 생성자: set()
  * 리스트나 튜플의 중복 요소 제거
* 요소 추가
  * 집합명.add(요소):하나의 요소 추가
  * 집합명.update(추가집함): 여러 요소를 추가
* 요소 삭제
  * 집합명.remove(요소값): 하나의 요소 삭제
  * 집합명.clear(): 모든 요소 삭제
* 집합연산
  * 교집합: &
  * 합집합:|
  * 차집합: -

```python
# 집합 요소 추가
a={1,2,3,4}        #튜블
b=[10, 20, 30, 30] #리스트
a=set(a)
b=set(b)
print(a)
print(b)

a.add(100)
b.update([100, 200, 300])
print(a)
print(b)

#집합요소 삭제
a.remove(100)
print(a)

b.clear()
print(b)
```

```python
# 집합 연산
a=(1,2,2,30)   #튜풀
b=[10,20,30,30]#리스트
a=set(a)
b=set(b)
print(f'a집합={ a }. b집합={ b }')
print(f'교집합 a&b={ a & b }')
print(f'교치합 a&b={a | b}')
print(f'차집합 a&b={a - b}')
```

```python
# 해결 문제1
# set으로 출입 명부 만들기
vip_names=['Newton', 'Kepler', 'Euler','Galileo']
room1_list=['Newton']
room2_list=['Euler', 'Galileo']

# 방에 있는 사람 모두를 출력하시오
# 오지 않은 사람을 출력하시오
room1_list=set(room1_list) 
room2_list=set(room2_list)
vip_names = set(vip_names)
print(f'{room1_list | room2_list}')
print(f'{vip_names - (room1_list | room2_list)}')
```
