출력과 입력 그리고 변수

```python
# 출력 : print() 함수로 출력하기

print(20*2)
print("hello world")
print('hello' , 'world') # 문자를 사이 공백을 두고 출력된다
print('hello'+'wrold') #문자 를 합 친다
```

```python
# 변수

a='hello world'

print(type(a)) #자료형 문자
print(a)

b=2
print(type(b)) #자료형 정수
print(b)
```

```python
#입력  input() 함수로 문자열 값 입력받기

name = input()
print(name+'님 ! 안녕하세요')
print(name,'님 ! 안녕하세요')
print(f'{name}님 ! 안녕하세요')
name='{}님! 안녕하세요'
print(name.format('심명효'))
```

```python
#입력 : 개선

age=int(input('나이를 입력해 주세요: '))
print(age-1)
```

```python
# 산술 연산자

print(10*2)
print(10**2)
print(10%2)
print(7/2)
print(7//2) #소수점 없이 나눔
```

```python
# 비교 연산자 

print(10>=2) 
print(10<=2)
print(10==2)
print(10!=2)
print(10%2==1) 
```

```python
#논리 연산자 (참: true, 거짓 False)
# and(둘 다 참일때 True), or(둘 중하나만 참일 때 true), not(반전)

#and 연산자
print(1==1 and 2!=1)
print(10%2!=0 and 1+1>0)
print()

# or 연산자
print(10<5 or 10==5)
print(10%2 !=0 or 1+1 >0)
print()

#not 연산자
print(not 10>5) 
print(not 10==5)
print(not 0)
print(not 10)
```

```python
# random 라이브러리 사용

import random
dice=random.randint(1,6)
print(dice)
```

```python
# for 반복문 - 1

name ='소마고'

for i in name:
  print(i,end="")
```

```python
#비밀번호 확인
passwd = int(input('비밀번호 4자리를 입력하세요:'))
if passwd == 2020:
   print('비밀번호가 일치합니다')
```

```python
# 비밀번호 해킹 프로그램 (for문과 if 문의 융합)

for i in range(1,100000):
  if i == 2023:
    print()
    print('비밀번로가 일치합니다')
    print(f'비밀번호는 {i}입니다')
```

```python
# else 문 추가
passwd=int(input('비밀번호 4자리를 입력하세요:'))
if passwd == 2020:
  print('비밀번호가 일치합니다')
else:
  print('비밀번호가 일치하지 않습니다')
```

```python
print('소름끼치도록 놀라운 심리테스트')

menu = input('당신이 좋아하는 음식을 입력해주세요')

if menu == '짜장면':
  print('당신은 짜장면을 좋아합니다')
elif menu == '떡복이':
  print('당신은 떡복이를 좋아합니다')
elif menu == '아이스크림':
  print('당신은 아이스크림을 좋아합니다')
else:
  print('당신은 짜장면, 아이스크림, 사탕을 좋아하지 않는 사함입니다')
  print('당신은 {}을 좋아합니다',format(menu))
```
