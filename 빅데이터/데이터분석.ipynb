```python
#에러 해결
filename = '/content/drive/MyDrive/DATA/busan.csv'

import csv
f = open(filename, 'r', encoding='cp949')
data = csv.reader(f) #콤마 
header=next(data)
save=0.2
save_data=[]
for row in data:
  if row[2] == '': #공백 처리
    row[2]=-99
  if float(row[2]) > float(save):
    save = row[2]
    save_data =row
print(save_data)
f.close()
```

```python
# 부산의 기온이 가장 낮았던 날의 날짜와 기온 구하기
filename = '/content/drive/MyDrive/DATA/busan.csv'

import csv
f = open(filename, 'r', encoding='cp949')
data = csv.reader(f) #콤마 
header=next(data)
save=0.2
save_data=[]
for row in data:
  if row[4] == '': #공백 처리
    row[4]=-99
  if float(row[4]) < float(save):
    save = row[4]
    save_data =row
print(save_data[0], end=',')
print(save, end='')
f.close()
```

```python
#데이터 시각화 하기
filename = '/content/drive/MyDrive/DATA/busan.csv'

import csv
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 
plt.title('마커 모양 변경')

f = open(filename, 'r', encoding='cp949')
data = csv.reader(f)
next(data)
print(data)
result=[] # 최고 기온 대이터를 저장할 리스트 생성
result2=[]
for row in data:
  if row[2] != '': #데이터가 없을 때
    result2.append(int(row[0][5:7]))
    result.append(float(row[2]))
f.close
print(result)
print(result2)
plt.plot(result2,result, 'r.',label="원(.)")
plt.legend() #범례 표시 함수
plt.show()
```

```python
# split()함수를 사용해 1년 중 여름의 정점인 8월의 최고 기온 데이터만을 추출해서 그래프로 그리시오
filename = '/content/drive/MyDrive/DATA/busan.csv'

import csv
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 
plt.title('마커 모양 변경')

f = open(filename, 'r', encoding='cp949')
data = csv.reader(f)
next(data)
print(data)

result=[] # 최고 기온 대이터를 저장할 리스트 생성

for row in data:
  if row[2] != '': #데이터가 없을 때
    rdata = row[0] 
    if row[0].split('-')[1] == '08':
      result.append(float(row[2]))
f.close
print(result)
plt.plot(result, 'r',label="원(.)")
plt.legend() #범례 표시 함수
plt.show()
```
