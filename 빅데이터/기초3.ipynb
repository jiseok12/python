컬렉션 데이터 활용
  * 컬렉션 데이터 모든 요소에 접근
  * for 요소변수 in 컬렉션명:
  * zip() 함수: 동일한 개수로 이루어진 자료형을 묶어 주는 역할을 하는 함수

```python
# 컬렉션 요소 접근
lst = [10,20,30]
tp=('a','b','c')

for item in lst:
  print(item,end=' ')
print()
for item in tp:
  print(item,end=' ')
```

```python
 # 딕셔너리 만들기: dict()함수와 zip()함수 이용
dt = dict(zip(tp,lst))
print(dt)
for item in dt:
  print(f'{item}:{dt[item]}')
```

### 내장함수
  * len(컬렉션명):컬렉션 요소 개수
  * max(컬렉션명):가장 큰 요소값 반환
    * 딕셔너리인 경우는 키 값 중 가장 큰 요소 반환
  * min(컬렉션명):가장 작은 요소값 반환
    * 딕셔너리인 경우는 키 값 중 가장 작은 요소 반환
  * sum(컬렉션명):수치 요소로 이루어진 컬렉션 요소의 합
    * 딕셔너리인 경우는 키 값의 합
  * sorted(컬렉션명): 컬렉션 요소 정렬
    * 딕션너리인 경우는 키 값을 정렬하여 키 리스트 반환

```python
# 내장함수

lst = [10,50,30]
print(len(lst))    #요소 개수
print(max(lst))    #최대값
print(min(lst))    #최소값ㄹ
print(sum(lst))    #합계
print(sorted(lst)) #정렬
print(lst)

lst.sort()
print(lst)
```
